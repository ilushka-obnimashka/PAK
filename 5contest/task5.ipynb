```python
df2 = pd.read_csv('wells_info_na.csv')
columns = df2.columns
series = pd.Series()
float_dict = {}
string_dict = {}
df2
```

```python
def median (x : list) -> float:
    n=len(x)
    if n%2 == 1 :
        return x[n//2]
    else :
        return 0.5*(x[n//2] + x[n//2 - 1])
```

```python
for i in columns:
    if (df2[i].dtype == 'float64'):
        freq_series =  df2[i].value_counts()
        temp_dict = freq_series.to_dict()
        float_dict = {**float_dict, **temp_dict} 
        
    if (df2[i].dtype == 'object'):
        freq_series =  df2[i].value_counts()
        temp_dict = freq_series.to_dict()
        string_dict = {**string_dict, **temp_dict} 


sorted_float_dict = {k: float_dict[k] for k in sorted(float_dict.keys())}
sorted_string_dict = {k: string_dict[k] for k in sorted(string_dict.keys(), key = string_dict.get,  reverse=True)}

max_key_string = max(sorted_string_dict, key=sorted_string_dict.get)


total_num = [x for x in sorted_float_dict for _ in range (sorted_float_dict[x])]
total_mean = median(total_num)

for i in columns:
    if (df2[i].dtype == 'float64'):
        df2[i] = df2[i].fillna(total_mean)
    if (df2[i].dtype == 'object'):
        df2[i] = df2[i].fillna(max_key_string)
```
