## Task1

```python
def palindrome_check(str1):
    
    str1 = str1.replace(' ','')
    str1 = str1.lower()
    
    for i in range(len(str1)):
        if (str1[i] == str1[len(str1)-i-1]):
            continue
        else:
            print ("string is not a palindrome")
            return
    print("string is a palindrome")
    return 
    
s = (input())
palindrome_check(s)
```

#### 1way

```python
def search_longest_word(str1):
    
    maxlen, currentlen, startidx = 0, 0, 0
    for i in range(len(str1)):
        if str1[i] == ' ':
            if currentlen > maxlen:
                maxlen = currentlen
                startidx = i - currentlen
                currentlen = 0
            else:
                currentlen = 0
        else:
            currentlen += 1

    if currentlen > maxlen:
        maxlen = currentlen
        startidx = i - currentlen + 1
        currentlen = 0
    return startidx


def print_longest_word(str1, idx):
    for i in range(idx, len(str1), 1):
        if str1[i] == ' ':
            return
        else:
            print(str1[i], end='')

s = input()
wordidx = search_longest_word(s)
print_longest_word(s, wordidx)
```

#### 1way

```python
from functools import lru_cache
from sys import maxsize


@lru_cache(maxsize = None)

def fib (num) :
    if num == 0 or num == 1 :
        return 1
    else :
        return fib(num-2) + fib(num-1)

start = time.perf_counter()
N = int(input())
print(fib(N), time.perf_counter() - start)
```

#### 2way

```python
M = { 0:1, 1:1}

def fib (num) :
    if num in M :
        return M[num]
    else :
        M[num] = fib(num-2) + fib(num-1)

start = time.perf_counter()
N = int(input())
fibnum = [fib(i) for i in range(N+1)]
print(fib(N), time.perf_counter() - start)
```

#### Формула для \\( n \\)-го члена геометрической прогрессии

\\(a_n\\) = \\( a_1 \\) * \\( r \\) ^ (\\(\\n\\)-1)

  где:
   - \\( a_n \\) — \\( n \\)-й член прогрессии,
   - \\( a_1 \\) — первый член прогрессии,
   - \\( r \\) — знаменатель прогрессии (коэффициент),
   - \\( n \\) — номер члена.

```python

def geometric_progression ():
    n = 1
    while True:
        an = a1 * r **(n-1)
        yield an
        n += 1


a1 = int(input())
r = int(input())
gp = geometric_progression() 
for num in gp:
    print(num)
```
